# 2. Homoscedasticity ( Constant Variance ): 

The assumption of Homoscedasticity is crucial to linear regression models. 
Homoscedasticity describes a situation in which the error term or variance or the "noise" or random disturbance in the relationship between the independent variables and the dependent variable is the same across all values of the independent variable. 


In other words, there is a constant variance present in the response variable as the predictor variable increases. 
If the "noise" is not same across the values of an independent variable like the residual plot below,
we call that 'Heteroscedasticity'.it is the opposite of Homoscedasticity.

```python
#we can also check the residual plot, which tells us how is the error variance across the true line. 
Lets look at the 'residual plot' for independent variable 'GrLivArea and our target variable SalePrice' .
```

```python
plt.subplots(figsize = (8,7))
sns.residplot(df.GrLivArea, df.SalePrice)
```

```python
#The error plot shows that as GrLivArea value increases, the variance also increases
'the residual plot looks more like a funnel'.mean 'Heteroscedasticity'.
```

```python
## drop dependent feature from dataset
x=data_lasso.drop(['SalePrice'],axis=1)
```

```python
### Apply Feature Selection
# first, specify the Lasso Regression model,
# then select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then use the selectFromModel object from sklearn, which will select the features whose coefficients are non-zero
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))# remember to set the seed, the random state in this function
feature_sel_model.fit(x,y)
```

```python
feature_sel_model.get_support()
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = x.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((x.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
#print('features with coefficients shrank to zero: {}'.format(np.sum(sel_.estimator_.coef_ == 0)))
```

# Data Modeling with selected feature

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns
```

# Preparing data for modeling

```python
x=X_train_lasso
y=data_lasso[['SalePrice']]
```

```python
#Take their values in X and y
X=x.values
Y= y.values
```

```python
#Split data into train and test formate
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=7)
```

# Linear Regression with lasso Features

```python
#Train the model
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
#Fit the model
model.fit(X_train, y_train)
```

```python
#Prediction
# predicting the values.
predictions = model.predict(X_test)
print("Predict value " + str(model.predict([X_test[142]])))
print("Real value " + str(y_test[142]))
```

```python
#Score/Accuracy
print("Accuracy --> ", model.score(X_test, y_test)*100)
```

```python
# Getting the model eveluation from predicted values
from sklearn.metrics import mean_squared_error, r2_score
# model evaluation
rmse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(rmse)
print(r2)
```
